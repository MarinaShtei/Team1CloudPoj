```python
import string
import json
from IPython.display import clear_output
from IPython.display import display
import ipywidgets as widgets
from google.colab import drive

drive.mount('/content/drive')
student_view = []
students = None


def read_students():
  with open("/content/drive/MyDrive/cloud/student.json", "r") as fid: 
    students = json.load(fid)
  for student in students:
    student_view.append(student["fname"] +" "+ student["lname"])
  return students

def write_students():
    student = get_selected()
    if Show.value != "":
      student["show"] = Show.value
    with open("/content/drive/MyDrive/cloud/student.json", "w") as fid: 
      json.dump(students,fid)

def reload():
  clear_output()
  print("Choose student:")
  display(Dropdown_)

def get_selected():
  return students[Dropdown_.index]

def display_courses(student):
    print("Courses:")
    for i,course in enumerate(student["courses"]):
      display(widgets.Text(course,disabled=True))

def display_mail(student):
    print("Mail:")
    Mail = widgets.Text(student["mail"],disabled=True)
    display(Mail)

def display_link(student):
    print("Link:")
    Link = widgets.Text(student["link"],disabled=True)
    display(Link)

def display_save_edit():
    if Show.disabled:
      display(Edit_btn)
    else:
      display(Save_btn)

def display_show(student):
    print("Favorite show:")
    Show.value = student["show"]
    display(Show)

def display_student(change):
    reload() #clear screen
    student = get_selected()
    display_mail(student)
    display_courses(student)
    display_link(student)
    display_show(student)
    display_save_edit()

students = read_students()
Show = widgets.Text(disabled=True)

Dropdown_ = widgets.Dropdown(
    options=student_view
)
Edit_btn = widgets.Button(
    icon='Edit',
    description='Edit favorite show'
)
Save_btn = widgets.Button(
    icon='Save',
    description='Save favorite show'
)

@Edit_btn.on_click
def edit_student(action):
    Show.disabled = False
    display_student(None)

@Save_btn.on_click
def save_student(action):
    write_students()
    Show.disabled = True
    display_student(None)

Dropdown_.observe(display_student, names='value')
output = widgets.Output()
display_student(None)
```
